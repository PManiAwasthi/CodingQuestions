<h3>Adding 2 polynomial equations</h3>

```python
#include<iostream>

struct Node{
    int cons;
    int expo;
    Node* next;
    
    Node(int constant, int exponent){
        cons = constant;
        expo = exponent;
        next = nullptr;
    }
};

Node* insertAtEnd(Node* head, int constant, int exponent){
    Node* temp = new Node(constant, exponent);
    
    if(head == nullptr || temp->expo>=head->expo){
        temp->next = head;
        head = temp;
        return head;
    }
    
    Node* curr = head;
    Node* prev = nullptr;
    while(curr!=nullptr && temp->expo<curr->expo){
        prev = curr;
        curr = curr->next;
    }
    temp->next = curr;
    prev->next = temp;
    
    return head;
}



Node* makeList(Node* head, int size){
    if(size==0){
        return head;
    }
    int constant, exponent;
    while(size!=0){
        std::cout<<"Enter value for the constant term:";
        std::cin>>constant;
        std::cout<<"Enter the exponent value:";
        std::cin>>exponent;
        head = insertAtEnd(head, constant, exponent);
        size--;
    }
    return head;
}

Node* addEquations(Node* equation1, Node* equation2){
    Node* result = nullptr;
    
    if(equation1 == nullptr && equation2 == nullptr){
        return result;
    }
    Node* ptr1 = equation1;
    Node* ptr2 = equation2;
    int sum = 0;
    int expo = 0;
    while(ptr1!=nullptr && ptr2!=nullptr){
        if(ptr1->expo == ptr2->expo){
            sum = ptr1->cons + ptr2->cons;
            expo = ptr1->expo;
            ptr1 = ptr1->next;
            ptr2 = ptr2->next;
        }else if(ptr1->expo < ptr2->expo){
            sum = ptr2->cons;
            expo = ptr2->expo;
            ptr2 = ptr2->next;
        }else if(ptr1->expo > ptr2->expo){
            sum = ptr1->cons;
            expo = ptr1->expo;
            ptr1 = ptr1->next;
        }
        result = insertAtEnd(result, sum, expo);
        sum = 0;
        expo = 0;
    }
    while(ptr1!=nullptr){
        result = insertAtEnd(result, ptr1->cons, ptr1->expo);
        ptr1 = ptr1->next;
    }
    while(ptr2!=nullptr){
        result = insertAtEnd(result, ptr2->cons, ptr2->expo);
        ptr2 = ptr2->next;
    }
    return result;
}

void printList(Node* head){
    if(head == nullptr){
        std::cout<<"List is empty..";
        return;
    }
    Node* ptr = head;
    while(ptr != nullptr){
        std::cout<<ptr->cons<<"X^"<<ptr->expo;
        if(ptr->next != nullptr){
            std::cout<<"+";
        }else{
            std::cout<<"= 0"<<std::endl;
        }
        ptr = ptr->next;
    }
}

int main(){
    Node* equation1 = nullptr;
    Node* equation2 = nullptr;
    
    //to input a diffrent equation uncomment this part
    // int size;
    // std::cout<<"Enter the number of terms in the equation1:";
    // std::cin>>size;
    // equation1 = makeList(equation1, size);
    // printList(equation1);

    // std::cout<<"Enter the number of terms in the equation1:";
    // std::cin>>size;
    // equation2 = makeList(equation2, size);
    
    //to use predefined values
    equation1 = insertAtEnd(equation1, 4, 3);
    equation1 = insertAtEnd(equation1, 12, 2);
    equation1 = insertAtEnd(equation1, 5, 1);
    equation1 = insertAtEnd(equation1, 5, 0);
    
    std::cout<<"equation1-> ";
    printList(equation1);
    
    equation2 = insertAtEnd(equation2, 3, 3);
    equation2 = insertAtEnd(equation2, 4, 2);
    equation2 = insertAtEnd(equation2, 2, 1);
    equation2 = insertAtEnd(equation2, 6, 0);
    
    std::cout<<"equation2-> ";
    printList(equation2);
    
    Node* result = addEquations(equation1, equation2);
    std::cout<<"result-> ";
    printList(result);
    return 0;
}
```

<h3>Multiplication of 2 equations</h3>

```python
#include <iostream>

struct Node{
    int coef;
    int expo;
    Node* next;
    
    Node(int constant, int exponent){
        coef = constant;
        expo = exponent;
        next = nullptr;
    }
};

Node* insertInOrder(Node* head, int constant, int exponent){
    Node* temp = new Node(constant, exponent);
    if(head == nullptr || temp->expo>head->expo){
        temp->next = head;
        return temp;
    }
    
    Node* curr = head;
    while(curr->next != nullptr && curr->next->expo>=temp->expo){
        if(curr->expo == temp->expo){
            curr->coef+=temp->coef;
            free(temp);
            return head;
        }
        curr = curr->next;
    }
    if(curr->expo==temp->expo){
        curr->coef+=temp->coef;
        free(temp);
    }else{
        temp->next = curr->next;
        curr->next = temp;
    }
    return head;
}

Node* make_list(Node* head, int size){
    if(size == 0){
        return head;
    }
    
    int constant = 0;
    int exponent = 0;
    
    while(size!=0){
        std::cout<<"Enter the constant term:";
        std::cin>>constant;
        std::cout<<"Enter the exponent term:";
        std::cin>>exponent;
        
        head = insertInOrder(head, constant, exponent);
        
        size++;
    }
    return head;
}

Node* multplyPolynomial(Node* equation1, Node* equation2){
    if(equation1 == nullptr){
        return equation2;
    }
    if(equation2 == nullptr){
        return equation1;
    }
    
    Node* ptr1 = equation1;
    Node* result = nullptr;
    while(ptr1 != nullptr){
        Node* ptr2 = equation2;
        while(ptr2 != nullptr){
            int constant = ptr1->coef * ptr2->coef;
            int exponent = ptr1->expo + ptr2->expo;
            result = insertInOrder(result, constant, exponent);
            ptr2 = ptr2->next;
        }
        ptr1 = ptr1->next;
    }
    return result;
}


void print(Node* head){
    if(head == nullptr){
        std::cout<<"List is empty";
        return;
    }
    Node* ptr = head;
    while(ptr!=nullptr){
        if(ptr->next!=nullptr){
            std::cout<<ptr->coef<<"X^"<<ptr->expo<<" + ";
        }else{
            std::cout<<ptr->coef<<"X^"<<ptr->expo<<" = 0";
        }
        ptr = ptr->next;
    }
    std::cout<<std::endl;
}

int main()
{
    Node* equation1 = nullptr;
    Node* equation2 = nullptr;
    int size;
    // std::cout<<"Enter the number of terms in the equation 1:";
    // std::cin>>size;
    
    // equation1 = make_list(equation1, size);
    equation1 = insertInOrder(equation1, 4, 3);
    equation1 = insertInOrder(equation1, 12, 2);
    equation1 = insertInOrder(equation1, 5, 1);
    equation1 = insertInOrder(equation1, 5, 0);
    
    std::cout<<"equation 1 -> ";
    print(equation1);
    
    // std::cout<<"Enter the number of terms in the equation 2:";
    // std::cin>>size;
    
    // equation2 = make_list(equation2, size);
    equation2 = insertInOrder(equation2, 3, 3);
    equation2 = insertInOrder(equation2, 4, 2);
    equation2 = insertInOrder(equation2, 2, 1);
    equation2 = insertInOrder(equation2, 6, 0);
    
    std::cout<<"equation 2 -> ";
    print(equation2);
    
    Node* result = multplyPolynomial(equation1, equation2);
    std::cout<<"result of multiplication-> ";
    print(result);
    

    return 0;
}
```

<h3>Adding 2 numbers</h3>

```python
#include <iostream>

struct Node{
    int digit;
    Node* next;
    
    Node(int d){
        digit = d;
        next = nullptr;
    }
};

Node* insertAtEnd(Node* head, int digit){
    Node* temp = new Node(digit);
    if(head == nullptr){
        return temp;
    }
    
    
    Node* ptr = head;
    
    while(ptr->next!=nullptr){
        ptr = ptr->next;
    }
    temp->next = ptr->next;
    ptr->next = temp;
    
    return head;
}

Node* insertAtFirst(Node* head, int digit){
    Node* temp = new Node(digit);
    if(head == nullptr){
        return temp;
    }
    Node* ptr = head;
    temp->next = ptr;
    head = temp;
    return head;
}

Node* make_list(Node* head, int num){
    int digit = 0;
    while(num!=0){
        digit = num%10;
        num = int(num/10);
        head = insertAtEnd(head, digit);
    }
    return head;
}

Node* reverse(Node* num){
   if(num == nullptr){
       return nullptr;
   }
   Node* prev = nullptr;
   Node* curr = num;
   Node* next = num;
   
   while(curr!=nullptr){
       next = curr->next;
       curr->next = prev;
       prev = curr;
       curr = next;
   }
   
   return prev;
}

Node* addition(Node* num1, Node* num2) {
    Node* result = nullptr;
    
    if(num1 == nullptr){
        return reverse(num2);
    }
    if(num2 == nullptr){
        return reverse(num1);
    }
    
    int sum;
    int carry = 0;
    while(num1 || num2){
        sum = 0;
        if(num1){
            sum += num1->digit;
            num1 = num1->next;
        }
        if(num2){
            sum += num2->digit;
            num2 = num2->next;
        }
        sum += carry;
        carry = sum / 10;
        sum = sum % 10;
        
        result = insertAtEnd(result, sum);
    }
    if(carry){
        result = insertAtEnd(result, carry);
    }
    return reverse(result);
}

void print(Node* head){
    if(head == nullptr){
        std::cout<<"List is empty.";
        return;
    }
    Node* ptr = head;
    while(ptr!=nullptr){
        std::cout<<ptr->digit;
        ptr = ptr->next;
    }
    std::cout<<std::endl;
}


int main()
{
    Node* num1 = nullptr;
    Node* num2 = nullptr;
    
    int number1, number2;
    
    std::cout<<"Enter the first number";
    std::cin>>number1;
    num1 = make_list(num1, number1);
    
    std::cout<<"Enter the second number";
    std::cin>>number2;
    num2 = make_list(num2, number2);
    
    Node* result = addition(num1, num2);
    std::cout<<"The result of addition is:";
    print(result);
    
    
    
    return 0;
}
```
