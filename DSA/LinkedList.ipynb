```python
#include <iostream>

struct Node{
    int value;
    Node* link;
    
    Node(int newdata){
        value = newdata;
        link=NULL;
    }
};

Node* insert_into_linked_list(Node* head){
    int value;

    
    Node* cur = head;

    std::cout<<"Enter the value of the element:";
    std::cin>>value;
    
    if(head==NULL){
       head = new Node(value);
       return head;
    }
    else{
        while(cur->link!=NULL){
        cur=cur->link;
        } 
    }

    Node* temp = new Node(value);
    cur->link = temp;
    
    return head;
}

Node* insert_given_position_pointer_linked_list(Node* head, int pos){
    int value;
    std::cout<<"Enter the value to insert:";
    std::cin>>value;
    Node* newNode = new Node(value);
    
    Node* prev;
    Node* next=head;
    if(next==NULL){
        return head;
    }else if(pos==1){
        newNode->link=head;
        head=newNode;
        return head;
    }else{
        while(pos!=1){
            prev = next;
            next = next->link;
            pos--;
        }
        if(next==NULL && pos>1){
            std::cout<<"List end reached";
        }else{
            newNode->link=next;
            prev->link=newNode;
        }
    }
    return head;
}

void print_list(Node* head){
    Node* ptr = head;
    while(ptr!= NULL){
        std::cout<<ptr->value;
        if(ptr->link!=NULL){
            std::cout<<"->";
        }
        ptr = ptr->link;
    }
}

int main()
{
    Node* head;
    int a = 1;
    int pos = 0;
    while(a){
        head = insert_into_linked_list(head);
        std::cout<<"\nWant to enter more element from the front of the list(0|no, 1|yes):";
        std::cin>>a;
        
    }
    std::cout<<"\nEnter the position where you want to insert the element:";
    std::cin>>pos;
    head = insert_given_position_pointer_linked_list(head, pos);
    print_list(head);
    
    return 0;
}
```

<h3>Reverse a linked list</h3>

```python
#include <iostream>

struct Node{
    int value;
    Node* link;
    
    Node(int newdata){
        value = newdata;
        link=NULL;
    }
};

Node* insert_into_linked_list(Node* head){
    int value;

    
    Node* cur = head;

    std::cout<<"Enter the value of the element:";
    std::cin>>value;
    
    if(head==NULL){
       head = new Node(value);
       return head;
    }
    else{
        while(cur->link!=NULL){
        cur=cur->link;
        } 
    }

    Node* temp = new Node(value);
    cur->link = temp;
    
    return head;
}

Node* reverse_linked_list(Node* head){
    Node *curr=head, *prev=NULL, *next;
    while(curr!=NULL){
        next = curr->link;
        curr->link=prev;
        
        prev = curr;
        curr = next;
        
    }
    return prev;
}

void print_list(Node* head){
    Node* ptr = head;
    while(ptr!= NULL){
        std::cout<<ptr->value;
        if(ptr->link!=NULL){
            std::cout<<"->";
        }
        ptr = ptr->link;
    }
}

int main()
{
    Node* head;
    int a = 1;
    int pos = 0;
    while(a){
        head = insert_into_linked_list(head);
        std::cout<<"\nWant to enter more element from the front of the list(0|no, 1|yes):";
        std::cin>>a;
        
    }
    head = reverse_linked_list(head);
    print_list(head);
    
    return 0;
}
```
