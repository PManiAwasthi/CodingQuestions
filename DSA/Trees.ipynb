<h3>Pre, In, post order traversal using recursion</h3>

```python
#include <iostream>

struct Node{
    int data;
    Node* left;
    Node* right;
    
    Node(){
        
    }
    
    Node(int newData){
        data = newData;
        left = nullptr;
        right = nullptr;
    }
    
    Node(int newData, Node* leftPtr, Node* rightPtr){
        data = newData;
        left = leftPtr;
        right = rightPtr;
    }
};

void preOrder(Node* root){
    if(root == nullptr){
        return;
    }
    std::cout<<root->data<<" ";
    
    preOrder(root->left);
    preOrder(root->right);
}

void inOrder(Node* root){
    if(root == nullptr){
        return;
    }
    inOrder(root->left);
    std::cout<<root->data<<" ";
    inOrder(root->right);
}

void postOrder(Node* root){
    if(root == nullptr){
        return;
    }
    postOrder(root->left);
    postOrder(root->right);
    std::cout<<root->data<<" ";
}

int main()
{
    Node* left = new Node(9);
    Node* right = new Node(20, new Node(15), new Node(7));
    Node* root = new Node(-10, left, right);
    
    std::cout<<"Preorder traversal: ";
    preOrder(root);
    std::cout<<"\nInorder traversal: ";
    inOrder(root);
    std::cout<<"\nPostorder traversal: ";
    postOrder(root);
    
    
    return 0;
}
```

<h3>Inorder traversal using iteration</h3>

```python
#include <iostream>

struct treeNode{
    int data;
    treeNode* left;
    treeNode* right;
    
    treeNode(int newData){
        data = newData;
        left = nullptr;
        right = nullptr;
    }
    
    treeNode(int newData, treeNode* leftPtr, treeNode* rightPtr){
        data = newData;
        left = leftPtr;
        right = rightPtr;
    }
}*root=nullptr, *left = nullptr, *right = nullptr;

struct Node{
    treeNode* data;
    Node* next;
    Node(treeNode* newData){
        data = newData;
        next = nullptr;
    }
}*top =nullptr;


int isEmpty(){
    if(top == nullptr){
        return 1;
    }
    return 0;
}

void push(treeNode* data){
    Node* temp = new Node(data);
    temp->next = top;
    top = temp;
}

treeNode* pop(){
    if(isEmpty()){
        std::cout<<"Underflow";
        return nullptr;
    }
    Node* temp = top;
    top = top->next;
    
    treeNode* data = temp->data;
    free(temp);
    return data;
}

void print(){
    if(top == nullptr){
        std::cout<<"List is empty.";
    }
    Node* ptr = top;
    while(ptr!=nullptr){
        std::cout<<ptr->data->data<<" ";
        ptr = ptr->next;
    }
    std::cout<<std::endl;
}

treeNode* peek(){
    return top->data;
}

void iterInOrder(){
    treeNode* curr = root;
    while(1){
        while(curr!=nullptr){
            push(curr);
            curr=curr->left;
        }
        if(top==nullptr){
            break;
        }
        curr = pop();
        std::cout<<curr->data<<" ";
        curr = curr->right;
    }
}

int main()
{
    left = new treeNode(9);
    right = new treeNode(20, new treeNode(15), new treeNode(7));
    root = new treeNode(-10, left, right);
    
    iterInOrder();
    return 0;
}
```

<h3>Preorder using iterations</h3>
