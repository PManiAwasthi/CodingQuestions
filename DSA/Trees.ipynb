<h3>Pre, In, post order traversal using recursion</h3>

```python
#include <iostream>

struct Node{
    int data;
    Node* left;
    Node* right;
    
    Node(){
        
    }
    
    Node(int newData){
        data = newData;
        left = nullptr;
        right = nullptr;
    }
    
    Node(int newData, Node* leftPtr, Node* rightPtr){
        data = newData;
        left = leftPtr;
        right = rightPtr;
    }
};

void preOrder(Node* root){
    if(root == nullptr){
        return;
    }
    std::cout<<root->data<<" ";
    
    preOrder(root->left);
    preOrder(root->right);
}

void inOrder(Node* root){
    if(root == nullptr){
        return;
    }
    inOrder(root->left);
    std::cout<<root->data<<" ";
    inOrder(root->right);
}

void postOrder(Node* root){
    if(root == nullptr){
        return;
    }
    postOrder(root->left);
    postOrder(root->right);
    std::cout<<root->data<<" ";
}

int main()
{
    Node* left = new Node(9);
    Node* right = new Node(20, new Node(15), new Node(7));
    Node* root = new Node(-10, left, right);
    
    std::cout<<"Preorder traversal: ";
    preOrder(root);
    std::cout<<"\nInorder traversal: ";
    inOrder(root);
    std::cout<<"\nPostorder traversal: ";
    postOrder(root);
    
    
    return 0;
}
```
