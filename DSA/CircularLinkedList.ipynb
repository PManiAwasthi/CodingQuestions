<h1>Circular Linked List</h1>
<h3>Definition</h3>
It is a type of linked list where the last element points at the first element of the list making it circular in nature. It can be implemented using a singly linked list or a doubly linked list. Another key difference in this data structure is that instead of a head pointer we use a tail pointer for all the operation on the linked list.

<h3>Insertion the first element</h3>

```python
#include <iostream>

struct Node{
    int data;
    Node* next;
    Node(int newdata){
        data = newdata;
        next = NULL;
    }
};

Node* addBeginning(Node* tail){
    int data;
    std::cout<<"enter the data value:";
    std::cin>>data;
    
    tail = new Node(data);
    tail->next = tail;
    
    return tail;
}

void printList(Node* tail){
    Node* ptr = tail->next;
    
    if(ptr==NULL){
        std::cout<<"\nNo element in the list.";
    }else{
        std::cout<<"\nThe elements of the list are: ";
        do{
            std::cout<<ptr->data<<" ";
            ptr = ptr->next;
        }while(ptr!=tail->next);        
    }

}

int main()
{
    char choice;
    std::cout<<"Do you want to create a circular linked list(y/n):";
    std::cin>>choice;
    
    Node* tail;
    
    if(choice=='y'||choice=='Y'){
        tail = addBeginning(tail);
    }else{
        std::cout<<"Thank you for using the program.";
        std::cout<<"";
    }
    
    printList(tail);
    

    return 0;
}
```

```python
#include <iostream>

struct Node{
    int data;
    Node* next;
    Node(int newdata){
        data = newdata;
        next = nullptr;
    }
};

Node* addBeginning(Node* tail, int data){
    
    tail = new Node(data);
    tail->next = tail;
    
    return tail;
}

Node* insertion_at_the_end(Node* tail, int data){
    
    Node* temp = new Node(data);
    
    if(tail == nullptr){
        tail = temp;
        tail->next = tail;
    }else{
        temp->next = tail->next;
        tail->next = temp;
        tail = temp;
    }
    
    return tail;
}

Node* make_list(int size){
    Node* tail=nullptr;
    
    if(size==0){
        return nullptr;
    }else{
        int data;
        while(size>=1){
            std::cout<<"Enter the data value:";
            std::cin>>data;
            tail = insertion_at_the_end(tail, data);
            size--;
        }
    }
    return tail;
}


int search_list(int data, Node* tail){
    if(tail == nullptr){
        std::cout<<"List is empty";
        return -1;
    }
    int flag = 0;
    Node* ptr = tail->next;
    int index = 0;
    do{
        if(ptr->data == data){
            flag = 1;
            break;
        }
        index++;
        ptr = ptr->next;
    }while(ptr!=tail->next);
    
    if(flag == 1){
        return index;
    }
    return -1;
}

void printList(Node* tail){
    if(tail==nullptr){
        std::cout<<"\nNo element in the list.";
        return;
    }
    
    Node* ptr = tail->next;
    std::cout<<"\nThe elements of the list are: ";
    do{
        std::cout<<ptr->data<<" ";
        ptr = ptr->next;
    }while(ptr!=tail->next);        

}

int main()
{
    char choice;
    int size, element;
    std::cout<<"Do you want to create a circular linked list(y/n):";
    std::cin>>choice;
    
    Node* tail=nullptr;
    
    if(choice=='y'||choice=='Y'){
        std::cout<<"enter the size of the list:";
        std::cin>>size;
        tail = make_list(size);
    }else{
        std::cout<<"Thank you for using the program."<<std::endl;
    }
    
    printList(tail);
    
    if(size!=0){
        std::cout<<"\nEnter the element to search for:";
        std::cin>>element;
        
        int index = search_list(element, tail);
        if(index!=-1){
            std::cout<<"\nElement found at: index = "<<index<<" and position = "<<++index;  
        }else{
            std::cout<<"\nElement not found";
        }
    }
    return 0;
}
```
