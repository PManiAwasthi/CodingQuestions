<h3>Evaluation of a postfix expression</h3>

```python
#include <iostream>
#include <string>

struct Node{
    int data;
    Node* next;
    
    Node(int newData){
        data = newData;
        next = nullptr;
    }
}*head = nullptr;

int isEmpty(){
    if(head == nullptr){
        return 1;
    }
    return 0;
}

void push(int data){
    Node* temp = new Node(data);
    temp->next = head;
    head = temp;
}

int pop(){
    if(isEmpty()){
        std::cout<<"Underflow.";
        return -1;
    }
    Node* temp = head;
    head = head->next;
    int data = temp->data;
    free(temp);
    return data;
    
}

void print(){
    if(head == nullptr){
        std::cout<<"List is empty.";
    }
    Node* ptr = head;
    std::cout<<std::endl;
    
    while(ptr!=nullptr){
        std::cout<<ptr->data<<" ";
        ptr = ptr->next;
    }
}

int calculate(char ch, int operand1, int operand2){
    int result = 0;
    switch(ch){
        case '+': result = operand2 + operand1;
        break;
        case '-': result = operand2 - operand1;
        break;
        case '*': result = operand2 * operand1;
        break;
        case '/': result = operand2 / operand1;
        break;
        case '^': result = operand2 ^ operand1;
        break;
    }
    return result;
}

int postfix_eval(std::string expression){
    int operand1, operand2, result;
    for(char ch:expression){
        operand1 = operand2 = 0;
        switch(ch){
            case '+':
            case '-':
            case '*':
            case '/':
            case '^':
            operand1 = pop();
            operand2 = pop();
            push(calculate(ch, operand1, operand2));
            break;
            default:
            push(ch-48);
        }
    }
    return pop();
}

int main()
{
    int result;
    std::string expression;
    
    std::cout<<"Enter the postfix expression to be evaluated:";
    std::cin>>expression;
    
    result = postfix_eval(expression);
    std::cout<<"The result is: "<<result;
    
    return 0;
}
```

<h3>Infix to Postfix</h3>

```python
#include <iostream>
#include <string>

struct Node{
    char data;
    Node* next;
    
    Node(char newData){
        data = newData;
        next = nullptr;
    }
}*head = nullptr;

int isEmpty(){
    if(head == nullptr){
        return 1;
    }
    return 0;
}

void push(int data){
    Node* temp = new Node(data);
    temp->next = head;
    head = temp;
}

char pop(){
    if(isEmpty()){
        std::cout<<"Underflow.";
        return -1;
    }
    Node* temp = head;
    head = head->next;
    char data = temp->data;
    free(temp);
    return data;
    
}

void print(){
    if(head == nullptr){
        std::cout<<"List is empty.";
    }
    Node* ptr = head;
    std::cout<<std::endl;
    
    while(ptr!=nullptr){
        std::cout<<ptr->data<<" ";
        ptr = ptr->next;
    }
}

char peek(){
    return head->data;
}

int precedence(char ch){
    switch(ch){
        case '^':return 3;
        break;
        case '*':
        case '/':
        return 2;
        break;
        case '+':
        case '-':return 1;
        default:return 0;
    }
}

std::string infixToPostfix(std::string expression){
    std::string result;
    char symbol;
    for(char ch:expression){
        switch(ch){
            case '+':
            case '-':
            case '*':
            case '/':
            case '^':
            symbol = ch;
            while(!isEmpty() && precedence(peek()) >= precedence(symbol)){
                result += pop();
            }
            push(symbol);
            break;
            case '(':
            push(ch);
            break;
            case ')':
            while(!isEmpty() && peek()!='('){
                result+=pop();
            }
            pop();
            break;
            default:
            result+=ch;
        }
    }
    while(!isEmpty()){
        result+=pop();
    }
    return result;
}

int main()
{
    std::string result;
    std::string expression;
    
    std::cout<<"Enter the infix expression to be evaluated:";
    std::cin>>expression;
    
    result = infixToPostfix(expression);
    std::cout<<result;
    return 0;
}
```
