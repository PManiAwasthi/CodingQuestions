<h3>Reverse a stack using using intermediate stacks such that the original stack is reversed.</h3>

```python
#include <iostream>

struct Node{
    int data;
    Node* next;
    Node(int newData){
        data = newData;
        next = nullptr;
    }
};

int isEmpty(Node* top){
    if(top == nullptr){
        return 1;
    }
    return 0;
}

void push(Node** top, int data){
    Node* temp = new Node(data);
    temp->next = *top;
    *top = temp;
}


void print(Node* top){
    if(top == nullptr){
        std::cout<<"\nList is empty.";
        return;
    }
    Node* temp = top;
    while(temp!=nullptr){
        std::cout<<temp->data<<" ";
        temp = temp->next;
    }
    std::cout<<std::endl;
}

int pop(Node** top){
    if(isEmpty(*top)){
        std::cout<<"Underflow";
        return -1;
    }
    Node* temp = *top;
    *top = (*top)->next;
    
    int data = temp->data;
    return data;
}

void reverseList(Node** top) {
    Node *temp1 = nullptr, *temp2 = nullptr;
    
    while(*top != nullptr){
        push(&temp1, pop(top));
    }
    
    while(temp1!=nullptr){
        push(&temp2, pop(&temp1));
    }
    
    while(temp2!=nullptr){
        push(top, pop(&temp2));
    }
}



int main()
{
    Node *top;
    
    top = nullptr;
    
    push(&top, 1);
    push(&top, 2);
    std::cout<<"Before reversing:";
    print(top);
    reverseList(&top);
    std::cout<<"After reversing:";
    print(top);
    
    return 0;
}
```
