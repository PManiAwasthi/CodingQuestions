<h1>Doubly Linked List</h1>
<h3><b>Definition</b></h3>
It is a type of linked list which uses node containing 2 pointers and a data block, and has a capability to move forward and backward without using excess of pointers that we need in case of single linked list.

<h3>Insertion from the end</h3>

```python
/******************************************************************************

                              Online C++ Compiler.
               Code, Compile, Run and Debug C++ program online.
Write your code in this editor and press "Run" button to compile and execute it.

*******************************************************************************/

#include <iostream>

struct Node{
    Node* prev;
    int data;
    Node* next;
    Node(int newData){
        prev = NULL;
        data = newData;
        next = NULL;
    }
};

Node* insertion_end(Node* head){
    int data, size;
    std::cout<<"Enter the size of the list";
    std::cin>>size;
    while(size!=0){
        std::cout<<"Enter the value to be inserted:";
        std::cin>>data;
        
        Node* temp = new Node(data);
        
        if(head==NULL){
            head = temp;
        }else{
            Node* ptr = head;
            while(ptr->next!=NULL){
                ptr=ptr->next;
            }
            ptr->next = temp;
            temp->prev = ptr;
        }   
        size--;
    }

    return head;
}

void print_list(Node* head){
    Node* ptr = head;
    while(ptr!=NULL){
        std::cout<<ptr->data;
        if(ptr->next!=NULL){
            std::cout<<"->";
        }
        ptr=ptr->next;
    }
}

int main()
{
    char choice;
    std::cout<<"Do you want to make a doubly linked list(y/n):";
    std::cin>>choice;
    Node* head;
    if(choice=='y'||choice=='Y'){
        head = insertion_end(head);
        print_list(head);
    }else{
        std::cout<<"Thank you for using the program";
    }
    
    return 0;
}
```

<h3>Deleting from a given position</h3>

<h3>Reversing a doubly linked list</h3>

```python
#include <iostream>

struct Node{
    Node* prev;
    int data;
    Node* next;
    Node(int newData){
        prev = NULL;
        data = newData;
        next = NULL;
    }
};

Node* insertion_end(Node* head){
    int data, size;
    std::cout<<"Enter the size of the list";
    std::cin>>size;
    while(size!=0){
        std::cout<<"Enter the value to be inserted:";
        std::cin>>data;
        
        Node* temp = new Node(data);
        
        if(head==NULL){
            head = temp;
        }else{
            Node* ptr = head;
            while(ptr->next!=NULL){
                ptr=ptr->next;
            }
            ptr->next = temp;
            temp->prev = ptr;
        }   
        size--;
    }

    return head;
}

Node* reversing_the_list(Node* head){
    Node* Prev;
    Node* Curr=head;
    Node* Next=head;
    
    if(Curr->next == NULL){
        return Curr;
    }
    
    while(Next->next!=NULL){
        Next = Next->next;
        Prev = Curr;
        Prev->next = Prev->prev;
        Curr = Next;
        Prev->prev = Curr;
    }
    Curr->next = Prev;
    Curr->prev = NULL;
    Prev = Curr;
    
    return Curr;
}

Node* reversing_using_recursion(Node* curr){
    if(curr == NULL){
        return NULL;
    }
    Node* temp = curr->prev;
    curr->prev = curr->next;
    curr->next = temp;
    if(curr->prev==NULL){
        return curr;
    }
    return reversing_using_recursion(curr->prev);
}



void print_list(Node* head){
    Node* ptr = head;
    std::cout<<"\nelement are:";
    while(ptr!=NULL){
        std::cout<<ptr->data;
        if(ptr->next!=NULL){
            std::cout<<"->";
        }
        ptr=ptr->next;
    }
}

int main()
{
    char choice;
    int data, pos;
    std::cout<<"Do you want to make a doubly linked list(y/n):";
    std::cin>>choice;
    Node* head;
    if(choice=='y'||choice=='Y'){
        head = insertion_end(head);
        print_list(head);
    }
    head = reversing_the_list(head);
    std::cout<<"\nAfter reversing normally:";
    print_list(head);
    head = reversing_using_recursion(head);
    std::cout<<"\nAfter reversing using recursion:";
    print_list(head);
    
    return 0;
}
```
